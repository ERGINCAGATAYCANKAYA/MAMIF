```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Training without obstacle 

### Size of cells 300 * 300
### Settings 
* GAMMA = 0.96  # discount rate
* LEARNING_RATE = 0.0001
* BATCH_SIZE = 300  # training batch size
* TARGET_UPDATE = 100  # number of moves to update the target network


* N_EPISODE = 30
* N_MOVES = 2000 # moves per episode


* REWARD_HIT_TREE = -1  # -1
* REWARD_MOVE = -0.3
* REWARD_BETTER_DISTANCE = 0.2
* REWARD_ON_EDGE = -0.9
* REWARD_GOAL = 10  # 5


```python
dfs = []
for i in range(1,8):
    file_path='../base_speed_'+str(i)+'.csv'
    df = pd.read_csv(file_path)
    dfs.append(df)

i = 1
dat = {}
for df in dfs:
    dat[i] = list(df.groupby('experiment')['reward'].sum())
    i += 1
cum_rewards = pd.DataFrame(dat)
cum_rewards.plot()
```
