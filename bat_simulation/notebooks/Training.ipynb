```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dfs = []
for i in range(1,6):
    file_path='../store/speed_'+str(i)+'.csv'
    df = pd.read_csv(file_path)
    dfs.append(df)

i = 1
dat = {}
for df in dfs:
    dat[i] = list(df.groupby('experiment')['reward'].sum())
    i += 1
cum_rewards = pd.DataFrame(dat)
```
